```python
import ollama
# 透過Ollama判斷是否為家具類別
def is_furniture_category(category_name):
    prompt = f"""
    Is the category '{category_name}' directly related to furniture?
    Answer with only 'Yes' or 'No'.
    """
    result = ollama.generate(model='deepseek-r1', prompt=prompt)
    return 'yes' in result['response'].lower()
```

```python
import json
import requests
from bs4 import BeautifulSoup
from pymongo import MongoClient
from playwright.sync_api import sync_playwright
import time

# 連線至 MongoDB（請根據實際情況調整連線字串）
client = MongoClient("mongodb://admin:zmxcnbv@localhost:27017/?authSource=admin")
db = client["furniture_db"]

# 假設 data 為你從 API 或其他來源取得的原始 JSON 二進位資料
# 例如:data = conn.getresponse().read()
json_data = json.loads(data.decode("utf-8"))

# 定義原始資料中欄位與目標欄位的對應關係
field_mapping = {
    "_id": "id",
    "image": "image",
    "name": "name",
    "url": "url",
    "description": "description",       # 原始資料中沒有此欄位
    "price": "price",                   # 這是一個物件
    "color_options": "color_options",   # 原始資料中沒有此欄位
    "view_url": "view_url",             # 原始資料中沒有此欄位
    "Image_matting": "Image_matting",   # 原始資料中沒有此欄位
    "images": "images",                 # 原始資料中沒有此欄位
    "price_string": "price_string",     # 原始資料中沒有此欄位
    "size_options": "size_options",     # 原始資料中沒有此欄位
    "asin": "asin",                     # 原始資料中沒有此欄位
    "star": "star",                     # 原始資料中沒有此欄位
    "brand": "brand",                   # 原始資料中沒有此欄位
    "starNum": "starNum",               # 原始資料中沒有此欄位
    "type": "typeName"
}

def fetch_missing_fields(url):
    result = {}
    with sync_playwright() as p:
        # 使用 Chromium 並設置 headless 模式
        browser = p.chromium.launch(headless=True)
        context = browser.new_context()
        page = context.new_page()
        page.goto(url)
        page.wait_for_timeout(2000)  # 等待網頁初步載入,依實際情況調整等待時間

        # 嘗試點擊「Product details」按鈕以展開產品描述
        try:
            page.click("button.pip-list-view-item__action")
            page.wait_for_timeout(2000)  # 點擊後等待動態內容載入
        except Exception as e:
            print("點擊按鈕失敗或找不到按鈕:", e)

        # 取得點擊後的頁面原始碼並用 BeautifulSoup 解析
        html = page.content()
        soup = BeautifulSoup(html, 'html.parser')
        
        # 取得產品描述（使用 class 為 pip-product-details__container 裡的 <p> 標籤）
        container = soup.find("div", class_="pip-product-details__container")
        if container:
            paragraphs = container.find_all("p", class_="pip-product-details__paragraph")
            description = "\n".join([p.get_text(strip=True) for p in paragraphs])
            result["description"] = description
        else:
            result["description"] = ""
        
        # 補充其他欄位,依需求調整:
        result["view_url"] = url
        result["sorce"] = "IKEA官方"
        result["color_options"] = ""
        result["Image_matting"] = ""
        result["images"] = []
        result["price_string"] = ""
        result["size_options"] = ""
        result["asin"] = ""
        result["star"] = ""
        result["brand"] = ""
        result["starNum"] = ""
        result["category"] = ""
        
        browser.close()
    return result

extracted_items = []
for item in json_data:
    # 根據原始資料建立初步字典
    extracted = {}
    for target_key, source_key in field_mapping.items():
        extracted[target_key] = item.get(source_key, None)
    
    # 檢查哪些欄位缺失（None 或空字串皆視為缺失）
    missing_fields = [key for key, value in extracted.items() if value in (None, "")]
    
    # 若缺少欄位且該產品有 url,就利用爬蟲補充資料
    if missing_fields and extracted.get("url"):
        additional_data = fetch_missing_fields(extracted.get("url"))
        for field in missing_fields:
            # 僅更新爬蟲取得的非空資料
            if field in additional_data and additional_data[field]:
                extracted[field] = additional_data[field]
    extracted_items.append(extracted)

# 輸出結果（或依需求存入 MongoDB）
print(json.dumps(extracted_items, indent=4, ensure_ascii=False))
```
